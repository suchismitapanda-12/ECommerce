```python
#ATM System
class Account:
    def __init__(self, acc_no, name, pin, balance=0):
        self.acc_no = acc_no
        self.name = name
        self.pin = pin
        self.balance = balance
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited successfully.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
        elif amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully.")

    def check_balance(self):
        print(f"Available Balance: ₹{self.balance}")

    def transfer(self, target_account, amount):
        if amount <= 0:
            print("Invalid transfer amount.")
        elif amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            target_account.balance += amount
            print(f"₹{amount} transferred to {target_account.name}")
accounts = {
    1001: Account(1001, "Suchismita", 1234, 5000),
    1002: Account(1002, "Rahul", 2345, 3000),
    1003: Account(1003, "Priya", 3456, 7000)
}

def atm():
    print("Welcome to Python ATM")

    acc_no = int(input("Enter Account Number: "))
    pin = int(input("Enter PIN: "))

    if acc_no in accounts and accounts[acc_no].pin == pin:
        user = accounts[acc_no]
        print(f"\n Welcome {user.name}!")

        while True:
            print("\n1. Check Balance")
            print("2. Deposit")
            print("3. Withdraw")
            print("4. Transfer")
            print("5. Exit")
            choice = input("Select option: ")

            if choice == '1':
                user.check_balance()
            elif choice == '2':
                amount = float(input("Enter amount to deposit: "))
                user.deposit(amount)
            elif choice == '3':
                amount = float(input("Enter amount to withdraw: "))
                user.withdraw(amount)
            elif choice == '4':
                target_acc = int(input("Enter target account number: "))
                if target_acc in accounts:
                    amount = float(input("Enter amount to transfer: "))
                    user.transfer(accounts[target_acc], amount)
                else:
                    print(" Target account not found.")
            elif choice == '5':
                print("Thank you for using ATM.")
                break
            else:
                print("Invalid option.")
    else:
        print("Invalid Account Number or PIN.")


# Run ATM
atm()
```

```python
#Student GradeBook
class Student:
    def __init__(self, name):
        self.name = name
        self.marks = []
    def add_marks(self, mark):
        self.marks.append(mark)
    def average(self):
        return sum(self.marks) / len(self.marks)
    def grade(self):
        avg = self.average()
        if avg >= 90:
            return "A"
        elif avg >= 75:
            return "B"
        elif avg >= 50:
            return "C"
        else:
            return "Fail"

# Create students
students = {}

while True:
    print("\n1. Add Student")
    print("2. Add Marks")
    print("3. View Result")
    print("4. Exit")
    choice = input("Enter choice: ")

    if choice == '1':
        name = input("Enter student name: ")
        students[name] = Student(name)
        print("Student added!")
    elif choice == '2':
        name = input("Enter student name: ")
        if name in students:
            mark = float(input("Enter mark: "))
            students[name].add_marks(mark)
            print("Mark added!")
        else:
            print("Student not found!")
    elif choice == '3':
        name = input("Enter student name: ")
        if name in students:
            student = students[name]
            print("Average:", student.average())
            print("Grade:", student.grade())
        else:
            print("Student not found!")

    elif choice == '4':
        break
    else:
        print("Invalid choice!")
```

```python
#Transactions
bal = 0
trans = []

while True:
    print("\n1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Transaction History")
    print("5. Exit")

    choice = int(input("Enter your choice = "))

    if choice == 1:
        amt = int(input("Enter the amount you want to deposit = "))
        bal += amt
        trans.append(f"Deposit: {amt}")
        print("Amount deposited successfully.")

    elif choice == 2:
        amt = int(input("Enter the amount to withdraw = "))
        if amt <= bal:
            bal -= amt
            trans.append(f"Withdraw: {amt}")
            print("Amount withdrawn successfully.")
        else:
            print("Insufficient balance.")

    elif choice == 3:
        print("Current Balance =", bal)

    elif choice == 4:
        if not trans:
            print("No transactions yet.")
        else:
            print("Transaction History:")
            for t in trans:
                print(t)

    elif choice == 5:
        print("Thank you! Exiting...")
        break

    else:
        print("Invalid choice. Please try again.")
```

```python
#Room Allocation
rooms = {
    101: [],
    102: [],
    103: []
}

MAX_CAPACITY = 2   # Maximum students per room

def allot_room(student_name):
    for room_no, students in rooms.items():
        if len(students) < MAX_CAPACITY:
            students.append(student_name)
            print(f"Room {room_no} allotted to {student_name}")
            return
    print("Sorry! All rooms are full.")

def view_rooms():
    print("\nRoom Allotment List:")
    for room_no, students in rooms.items():
        print(f"Room {room_no}: {students}")

while True:
    print("\n1. Add Student")
    print("2. View Room Allotment")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter student name: ")
        allot_room(name)

    elif choice == "2":
        view_rooms()

    elif choice == "3":
        print("Sorry! All the rooms are full")
        break

    else:
        print("Invalid choice! Try again.")
```
